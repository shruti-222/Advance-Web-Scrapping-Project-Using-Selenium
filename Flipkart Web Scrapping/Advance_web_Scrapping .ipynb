```python
with open('Flipkart1.html','r',encoding='utf-8') as f:
    html = f.read()
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
soup= BeautifulSoup(html,'lxml') #lxml parcel
```

```python
containers=soup.find_all('div',{'class':'_1AtVbE col-12-12' })
```

```python
Name = []
Price = []
Processor =[]
Memory = []
Other_Information =[]
for i in containers:
    div_elem = i.find('div',{'class':'_4rR01T'})
    if div_elem:
        Name.append(div_elem.text.split('-')[0].strip())
    else:
        Name.append(np.nan)
    div_tag = i.find('div', {'class': '_30jeq3 _1_WHN1'})
    if div_tag:
        Price.append(div_tag.text)
    else:
        Price.append(np.nan)
    li_procc= i.find('li',{'class':'rgWa7D'}) 
    if li_procc:
        Processor.append(li_procc.text)
    else :
        Processor.append(np.nan)
    li_ram= i.find('div',{'class':'rgWa7D'}) 
    if li_ram:
        print(li_ram.text)
    else :
        np.nan
    
    ul_ram = i.find('ul', {'class': '_1xgFaf'}) 
    if ul_ram:
        ram_text = ul_ram.text
        ram_values = []
        if '8 GB DDR4 RAM' in ram_text:
            ram_values.append('8 GB DDR4 RAM')
        if '16 GB DDR4 RAM' in ram_text:
            ram_values.append('16 GB DDR4 RAM')
        if ram_values:
            Memory.append(' '.join(ram_values))
        else:
            Memory.append(np.nan)
    else:
        Memory.append(np.nan)
    whole_container = i.find('ul', {'class': '_1xgFaf'})
    if whole_container:
        Other_Information.append(whole_container.text)
    else:
        Other_Information.append(np.nan)
```

```python
laptop = pd.DataFrame({
    'Name':Name,
    'Price':Price,
    'Processor':Processor,
    'Memory':Memory,
    'Other_Information':Other_Information
})
```
